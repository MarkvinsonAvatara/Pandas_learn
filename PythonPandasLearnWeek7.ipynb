# Pengaturan Tampilan

untuk mengatur tampilan kita hanya perlu menggunakan: 
- fungsi pd.set_option('pengaturan tampilan apa yang ingin mau kita lakukan', ukuran yang ingin kita tentukan)

```python
#Contoh
pd.set_option('display.max_rows', 7) # Menentunkan jumlah maximal baris 
pd.set_option('display.max_columns', 8) #Menentunkan jumlah maximal kolom
pd.set_option('display.max_colwidth', 25) #Menentukan jumlah maximal karkter untuk setiap kolom

pf
#note jika ada yang lebih dari nilai maximal yang ditetapkan akan diberi tanda "...""
```

```python
#contoh 2
pd.reset_option('^display.', silent=True) #mereset ulang display option ke kondisi default
pf
```
