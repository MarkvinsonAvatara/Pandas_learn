# 33. Nilai Kumulatif pada Pandas Data Frame

```python
#import modules
import pandas as pd
import numpy as np 

#check Version
print(pd.__version__)
print(np.__version__)
```

```python
#Data Frame Prep
d= {'Pemain':['Arcee','Jack','Alex','Arcee','Arcee','Alex','Nattu','Jack'],
   'goal':[2,1,3,1,1,2,2,3]}
df=pd.DataFrame(d)
df
```

# Menampilkan Nilai Kumulatif

Nilai Kumulatif adalah jumlah nilai saat ini akan ditambahkan dengan keseluruhan jumlah nilai sebelumnya

```python
df['goal'].cumsum().to_frame()
```

```python
df['jumlah_goal_kumulatif'] = df['goal'].cumsum()
df
```

```python
df['jumlah_goal_kumulatif_tiap_pemain']=df.groupby('Pemain')['goal'].cumsum()
df
```

```python
df['cummax']=df['goal'].cummax()
df
```

```python
df['cummin']=df['goal'].cummin()
df
```

```python
df['cumprod']=df['goal'].cumprod()
df
```

note:
- .cumsum(): mendata jumlah suatu kolom secara kumulatif
- .cummax(): mendata nilai tertinggi yang diperoleh sampai dengan titik tertentu
- .cummin(): Mencari nilai terendah secara kumulatif
- .cumpord(): hasil total perkalian secara kumulatif

# 34. Mapping Data Frame dengan applymap()

```python
#import modules
import pandas as pd
import numpy as np 

#check Version
print(pd.__version__)
print(np.__version__)
```

```python
#Data Frame Prep
af=pd.DataFrame({'gender':['Pria','Wanita','lelaki','Lelaki','perempuan'],
                'usia':[23,21,24,22,21],
                'shift':['pagi','siang','malam','siang','pagi']})
af
```

# Mapping pada Data Frame dengan applymap()

```python
af=af.applymap(lambda x: x.lower() if type(x) == str else x)
af
```

note:
- lamba x akan menerima nilai x dan menguji tipe datanya, apabila tipe datanya String maka nilai tersebut akan dijadikanlowercase tetapi datanya bkan tipe data String maka akan dikembalikan apa adanya.
- Dengan menggunakan applymap() ekspresi lamba x akan dikenakan ke seluruh data frame

# applymap() dengan dictionary

```python
mapping = {'pria':'L',
          'lelaki':'L',
          'wanita':'P',
          'perempuan':'P',
          'pagi':1,
          'siang':2,
          'malam':3}
#variable mapping digunakan sebagai aturan untuk menstandarisasi nilai-nilai yang ada di Data Frame
af.applymap(mapping.get)
#untuk menerapkan variable mapping ke seluruhan Data Frame kita cukup menggunakan fungsi diatas
```

karena variable mapping hanya diperuntukan untuk kolom gender dan shift saja maka kita dapat menggunakan cara dibawah ini agar kolom usia tidak terterapkan di aturan mapping

```python
af[['gender','shift']]= af[['gender','shift']].applymap(mapping.get)
af                                   
```

# 35. Memadukan fungsi agregasi dengan transform()

```python
#import modules
import pandas as pd
import numpy as np 

#check Version
print(pd.__version__)
print(np.__version__)
```

```python
#Data Frame Prep
d={'no_nota':[1,1,1,2,2,3,4,5],
  'kopi':['latte','cappuccino','espresso','latte','espresso','cappuccino','latte','espresso'],
  'harga':[50,60,80,150,120,60,100,40]}

tf=pd.DataFrame(d)
tf
```

# Menghitung total harga untuk tiap nomor nota

```python
tf.groupby('no_nota')['harga'].sum().to_frame()
```

```python
tf['total_harga']=tf.groupby('no_nota')['harga'].transform(sum)
tf
```

# Menghitung total omset untuk tiap jenis kopi yang terjual

```python
tf.groupby('kopi')['harga'].sum().to_frame()
```

```python
tf['total_omset']=tf.groupby('kopi')['harga'].transform(sum)
tf
```

note: method transform() adalah method yang digunakan agar tidak terjadi ketimpangan baris saat kita menghitung suatu nilai berdasarkan suatu kolom tertentu

# 36. Menyatukan Kolom dengan str.cat()

```python
#import modules
import pandas as pd
import numpy as np 

#check Version
print(pd.__version__)
print(np.__version__)
```

```python
#Data Frame Prep
data={'nama':['jack','arcee','orion','Azzy'],
     'gender':['L','L','P','L'],
     'usia':[23,21,22,25]}

cf=pd.DataFrame(data)
cf
```

# Menyatukan Kolom dengan str.cat()

```python
#contoh 1
cf['nama'].str.cat(cf['gender'], sep=',').to_frame()
```

```python
#contoh 2
cf['nama_gen']=cf['nama'].str.cat(cf['gender'], sep=',')
cf
```

```python
#contoh 3
cf['nama'].str.cat(cf['usia'].astype('str'),sep='-').to_frame()
```

```python
#contoh 4
cf['nama_usia']=cf['nama'].str.cat(cf['usia'].astype('str'),sep='-')
cf
```

note: paramter sep adalah parameter untuk memisahkan nilai dari kolom yang digabung (parameter ini optional)
